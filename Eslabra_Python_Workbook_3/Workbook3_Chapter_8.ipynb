Create a list containing the valid grade values, then using enumerate in a for loop, print out all indices and items in the list.

```python
0   1.0
1   1.25
2   1.5
3   1.75
4   2.0
5   2.25
6   2.5
7   3.0
```

```python
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]
for index, grade in enumerate(grades):
    print(index, grade)
```

Modify the code in the subheading Parsing Lines to display all the time stamps of the emails in mboxt-short.txt

```python
fhand = open('/content/drive/MyDrive/wb3/mbox-short.txt')
emails = []
timestamps = []

for line in fhand:
    line = line.strip()
    if line.startswith('From '):
        words = line.split()
        emails.append(words[1])
        timestamp = words[5] + ' ' + words[2] + ' ' + words[4] + ' ' + words[6]
        timestamps.append(timestamp)

for email, timestamp in zip(emails, timestamps):
    print("Email:", [email])
    print("Timestamp:", [timestamp])
```
