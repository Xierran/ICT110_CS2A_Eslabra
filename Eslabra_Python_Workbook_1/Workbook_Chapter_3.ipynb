### Exercise 2



```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    rp = 40 * rate
    op = (hours - 40) * (rate * 1.5)
    tp = rp + op
else:
    tp = hours * rate

print("Pay:", tp)
```

```python
score = int(input('Enter the score:'))

if 97 <= score <= 100:
    grade = 1.0
elif 93 <= score < 97:
    grade = 1.25
elif 89 <= score < 93:
    grade = 1.5
elif 85 <= score < 89:
    grade = 1.75
elif 81 <= score < 85:
    grade = 2.0
elif 77 <= score < 81:
    grade = 2.25
elif 74 <= score < 77:
    grade = 2.5
elif 0 <= score < 74:
    grade = 3.0
else:
    grade = None
if grade is not None:
    print(f' Grade: {grade}')
else:
    print('Invalid score entered.')
```
