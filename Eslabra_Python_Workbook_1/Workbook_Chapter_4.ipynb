### Exercise 1


```python
def compute_pay(hours, rate):
    if hours > 40:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        total_pay = regular_pay + overtime_pay
    else:
        total_pay = hours * rate
    return total_pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)
print("Pay:", pay)
```

### CHALLENGE


```python
def computegrade(score):
    if 97 <= score <= 100:
        return '1.0'
    elif 93 <= score < 97:
        return '1.25'
    elif 89 <= score < 93:
        return '1.5'
    elif 85 <= score < 89:
        return '1.75'
    elif 81 <= score < 85:
        return '2.0'
    elif 77 <= score < 81:
        return '2.25'
    elif 74 <= score < 77:
        return '2.5'
    else:
        return '3.0'


scores = [100, 95, 90, 87, 83, 79, 75, 70]
for score in scores:
    print("Score:", score, "Grade:", computegrade(score))
```
