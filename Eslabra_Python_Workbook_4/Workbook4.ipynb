How do you calculate logarithms? Logarithms base 10? Natural Logarithms?

Constants in the math module include:

math.pi: π (pi)

math.e: e (base of natural logarithm)

math.tau: τ (tau) which is 2π

math.inf: Positive infinity

math.nan: NaN (Not a Number)

math.exp(x) returns e raised to the power of x, equivalent to
e^x.

math.pow(x, y) returns x raised to the power of y.

Logarithms can be calculated using functions like:

math.log(x, [base]): Returns the logarithm of x to the given base (default base is e).

math.log10(x): Returns the base-10 logarithm of x.

math.log1p(x): Returns the natural logarithm of 1+x (accurate for x near zero).

math.log2(x): Returns the base-2 logarithm of x.


For logarithms base 10, use math.log10(x).

For natural logarithms (base e), use math.log(x).
